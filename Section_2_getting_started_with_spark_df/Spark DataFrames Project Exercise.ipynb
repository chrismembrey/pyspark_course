## Bonus Question!
#### There are too many decimal places for mean and stddev in the describe() dataframe. Format the numbers to just show up to two decimal places. Pay careful attention to the datatypes that .describe() returns, we didn't cover how to do this exact formatting, but we covered something very similar. [Check this link for a hint](http://spark.apache.org/docs/latest/api/python/pyspark.sql.html#pyspark.sql.Column.cast)

If you get stuck on this, don't worry, just view the solutions.

```python
from pyspark.sql.functions import format_number


# remember to always create a new variable for the df.describe() variable
summary_df = df.describe()

formatted_df = summary_df.select(
    [
        format_number(summary_df[col].cast("double"), 2).alias(col)
        if col != "summary" else summary_df[col]
        for col in summary_df.columns
    ]
)

formatted_df.show()
```

#### Create a new dataframe with a column called HV Ratio that is the ratio of the High Price versus volume of stock traded for a day.

```python
df.select(
(df['High']/df['Volume']).alias('HV Ratio')
).show()
```

#### What day had the Peak High in Price?

```python
from pyspark.sql.functions import max, min, mean, median

df.select('Date', 'High').orderBy(df['High'].desc()).head(1)[0].asDict()['Date']
```

#### What is the mean of the Close column?

```python
df.select(
    format_number(mean("Close"),2).alias("avg(Close)")
).show()
```

#### What is the max and min of the Volume column?

```python
df.select(
    min("Volume").alias("min vol"),
    max("Volume").alias("max vol")
).show()
```

#### How many days was the Close lower than 60 dollars?

```python
df.filter("Close < 60").count()
```

#### What percentage of the time was the High greater than 80 dollars ?
#### In other words, (Number of Days High>80)/(Total Days in the dataset)

```python
from pyspark.sql.functions import approx_count_distinct

decimal_val = (df.filter("High > 80").count()/approx_count_distinct("Date"))

df.select(format_number((decimal_val * 100),2).alias("percentage high > 80")).show()
```

#### What is the Pearson correlation between High and Volume?
#### [Hint](http://spark.apache.org/docs/latest/api/python/pyspark.sql.html#pyspark.sql.DataFrameStatFunctions.corr)

```python
from pyspark.sql.functions import corr


pearson_corr = df.select(corr("High", "Volume")).collect()[0][0]
print("Pearson correlation between High and Vol:", pearson_corr)
```

```python
df.select(corr('High', 'Volume').alias('Pearson Corr of High and Vol')).show()
```

#### What is the max High per year?

```python
from pyspark.sql.functions import dayofmonth, hour, dayofyear, month, year, weekofyear

high_date = df.select(
year("Date").alias("year"),
    "High"
)

high_date.show()
```

```python
high_date.groupBy("year").agg(
    format_number(max("High"), 2).alias('Max High')
).orderBy("year").show()
```

#### What is the average Close for each Calendar Month?
#### In other words, across all the years, what is the average Close price for Jan,Feb, Mar, etc... Your result will have a value for each of these months.

```python
close_df = df.select(
month("Date").alias("month"),
    "Close"
)

close_df.show()
```

```python
close_df.groupBy("month").agg(
    format_number(mean("Close"), 2).alias('Mean Close')
).orderBy("month").show()
```
